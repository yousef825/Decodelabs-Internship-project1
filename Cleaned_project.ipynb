## Step 3: Cleaning 

```python
# Convert Date column to datetime format
df_dataset['Date'] = pd.to_datetime(df_dataset['Date'])
```

```python
df_dataset['Year'] = df_dataset['Date'].dt.year
df_dataset['Month'] = df_dataset['Date'].dt.month
df_dataset.head(10)
```

```python
# Missing values percentage
missing_percentage = df_dataset.isnull().sum() / len(df_dataset) * 100
print(missing_percentage)
```

```python
# fill missing values in Couponcode column with 'no_coupon'
df_dataset['CouponCode'] = df_dataset['CouponCode'].fillna('no_coupon')
df_dataset['CouponCode'].value_counts()

# fill missing values in CustomerID column with 'Unknown'
df_dataset['CustomerID'] = df_dataset['CustomerID'].fillna('Unknown')
df_dataset['CustomerID'].value_counts()

df_dataset.head(10)
```

```python
# check duplicates
df_dataset.duplicated().sum()

# if there are duplicates, drop them
df_dataset.drop_duplicates(inplace = True)
```

```python
# check unique OrderIDs
df_dataset['OrderID'].nunique()
```

```python
# Calculate the expected total price (Quantity × UnitPrice)
df_dataset['Calculated_Total'] = df_dataset['Quantity'] * df_dataset['UnitPrice']

# Create a True/False column: True if mismatch exists, False if they match
df_dataset['Mismatch'] = (df_dataset['Calculated_Total'] != df_dataset['TotalPrice'])

# Count how many rows have mismatches
df_dataset['Mismatch'].sum()
```

```python
# check if there are still missing values
df_dataset.isnull().sum()
```

```python
#check the final cleaned dataset
df_dataset.head(10)
```

```python
# standardize PaymentMethod values
df_dataset['PaymentMethod'] =(
    df_dataset['PaymentMethod'].str.lower().str.strip()
)
```

```python
df_dataset.to_csv('Dataset/cleaned_dataset.csv', index = False)
```
